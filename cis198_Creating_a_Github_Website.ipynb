### github notes

key terms:
* **repository**- a container holding all *comitted* flies and folders.
* **commit**- having a file version in a repository (along with a save name, i.e commit message)
* **git**- version control program (i.e. an app for saving many versions of files, ex. computer code)
* **github**- a online company (website) providing a place for users to save their repositories.
